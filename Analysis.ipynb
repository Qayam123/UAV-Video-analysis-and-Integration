```python
from math import radians, degrees, atan2, sin, cos
import numpy as np
from datetime import datetime
import haversine as hs
import numpy as np

class uav:

    def __init__(self,lat1, lon1, lat2, lon2,t1,t2):
        self.lat1=lat1
        self.lon1=lon1
        self.lat2=lat2
        self.lon2=lon2
        self.t1=t1
        self.t2=t2
        
    def distance(self):
        loc1 = (self.lat1, self.lon1)
        loc2 = (self.lat2, self.lon2)
        dist1 = hs.haversine(loc1, loc2)
        return dist1


    def compute_speed(self,dist1):
   
        t1 = self.t1
        t2 = self.t2 
        time_diff_sec = (t2 - t1)
        print(f"time difference is:{time_diff_sec}")
        speed_kts = dist1*1944/time_diff_sec
        print(f"speed attained in kts is {speed_kts}")
        return speed_kts
    def compute_course(self):
   

    # Convert coordinates to radians
        lat1, lon1, lat2, lon2 = map(radians, [self.lat1, self.lon1, self.lat2, self.lon2])

    # Compute initial bearing
        dLon = lon2 - lon1
        y = sin(dLon) * cos(lat2)
        x = cos(lat1) * sin(lat2) - sin(lat1) * cos(lat2) * cos(dLon)
        bearing = degrees(atan2(y, x))
        init_bearing = (bearing + 360) % 360
        print(f"Initial bearing is {init_bearing}")

    # Compute Final bearing
        dLon = lon1 - lon2
        y = sin(dLon) * cos(lat2)
        x = cos(lat2) * sin(lat1) - sin(lat2) * cos(lat1) * cos(dLon)
        bearing = degrees(atan2(y, x))
        bearing = (bearing + 360) % 360
        final_bearing = (bearing + 180) % 360
        print(f"final bearing is {final_bearing}")
        course = np.round((final_bearing - init_bearing), 4)
        return course
        print(f"course is {course}")

if __name__=="__main__":
    uav_obj=uav(5.317716,86.384318,5.317684,86.384427,1679055514,1679055516)
    uav_obj.compute_speed(uav_obj.distance())
    uav_obj.compute_course()
```

```python
import numpy as np
import cv2
cap = cv2.VideoCapture(0)
while(True):
# Capture frame-by-frame
    ret, frame = cap.read()
# Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
# Display the resulting frame
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
